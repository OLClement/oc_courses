```python
# sunspot 
sunspot = pd.read_csv('datasets/sunspot.year.csv',index_col='t')
sunspot.plot(y="sunspot", figsize=(14,4))
plt.show()
```

```python
np.random.seed(1789)
norm = pd.DataFrame({'Bruit blanc gaussien de variance 9': np.random.normal(scale=3, size=100), 't': range(100)})
norm.set_index('t')

norm.plot(y="Bruit blanc gaussien de variance 9", figsize=(14,4))
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```

```python
# uspop
uspop = pd.read_csv('datasets/uspop.csv',index_col='t')
uspop.plot(y='uspop', figsize=(14,4))
plt.show()
```

```python
# airpass
airpass = pd.read_csv('datasets/AirPassengers.csv',index_col='t')
airpass.plot(y='Airpass',figsize=(14,4))
np.log(airpass).plot(y='Airpass',figsize=(14,4))
plt.show()
```
